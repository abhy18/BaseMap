```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
```

```python
fig = plt.figure(figsize=(8,8))
m=Basemap()
m.drawcoastlines()
```

```python
m = Basemap(projection='ortho',lat_0=50,lon_0=80)
m.drawmapboundary(fill_color='aqua')
m.fillcontinents(color='coral',lake_color='aqua')
m.drawcoastlines()
plt.show()
```

```python
fig = plt.figure(figsize=(8,8),edgecolor='w')
m = Basemap(projection='cyl',llcrnrlat=-50,urcrnrlat=40,
            llcrnrlon=-60, urcrnrlon=60,)
m.drawcoastlines()
plt.show()
```

```python
fig = plt.figure(figsize=(8,8),edgecolor='w')
m = Basemap(projection='cyl',llcrnrlat=-50,urcrnrlat=40,
            llcrnrlon=-60, urcrnrlon=60,)
m.shadedrelief(scale=0.1)
```

```python
fig = plt.figure(figsize=(8,8))
m = Basemap(projection='lcc',width=15E6,height=16E6,
            lat_0=20,lon_0=77)
m.etopo(scale=0.5,alpha=0.5)

x,y = m(80.80000000,22.09000000)
plt.plot(x,y,'ok',markersize=5)
plt.text(x,y,'India',fontsize=12);
```
